## **Procesamiento de lenguaje natural (NLP)**

Análisis de texto y generación de lenguaje a partir de un corpus literario. Se requiere tanto la extracción de patrones lingüísticos en el texto como la generación de nuevo contenido en el estilo del autor elegido.

Autor: Edgar Allan Poe

**Nota 2:** La parte de código que resuelve los literales a) y b) es la parte que dice "Solución de los puntos a y b", las otras partes de códigos son pruebas que aunque contienen procesos iguales solo son código que quiero tener guardado.


*El objetivo del código es mostrar una gráfica que muestra la frecuencia de los 5 n-gramas más comunes del corpus. Además, usando el corpus se genera un texto aleatorio usando el mismo estilo del autor.*

#### Evitando palabras del archivo function_words.txt

```python
import nltk
from collections import Counter
import random
import matplotlib.pyplot as plt

def main():
  # Cargar datos
  print("Cargando datos...")
  textos_gutenberg = ["Barrica", "El_Cuervo", "Las_Campanas", "Muerte_Roja"]
  corpus = [load_text(f'{texto}.txt') for texto in textos_gutenberg]

  # Cargar palabras comunes a evitar
  function_words = load_text('function_words.txt').split()

  # Calcular los n-gramas y graficar los n-gramas más comunes para cada texto
  n = 2  # Escoger el valor de n-gramas según la preferencia
  for i, texto in enumerate(textos_gutenberg):
    print(f"\nProcesando {texto}...")
    filtered_corpus = filter_common_words(corpus[i], function_words)
    ngrams = Counter(get_ngrams([filtered_corpus], n))
    plot_ngram_frequency(ngrams, title=f'{n}-gramas más comunes para {texto}')

    # Generar texto aleatorio en el estilo del autor para cada texto
    longitud_texto = 50  # Escoger la longitud del texto generado para cada texto
    random_text = generate_random_text([filtered_corpus], n, length=longitud_texto)
    print(f'\nTexto Aleatorio Generado para {texto}:')
    print(random_text)

def load_text(file_name):
  with open(file_name, 'r', encoding='utf-8') as file:
    text = file.read()
  return text

def get_ngrams(files, n):
  ngrams = []
  for file in files:
    words = file.split()
    ngrams.extend(tuple(words[i:i + n]) for i in range(len(words) - n + 1))
  return ngrams

def filter_common_words(text, common_words):
  words = text.split()
  filtered_words = [word for word in words if word.lower() not in common_words]
  return ' '.join(filtered_words)

def plot_ngram_frequency(ngrams, title="Most Common n-grams"):
  combi = 5  # Escoger combinaciones de palabras más comunes
  most_common_ngrams = ngrams.most_common(combi)
  labels, values = zip(*most_common_ngrams)
  indexes = range(len(labels))

  # Modificar el tamaño de la gráfica
  plt.figure(figsize=(15, 3))

  plt.bar(indexes, values)
  plt.xticks(indexes, labels)
  plt.title(title)
  plt.show()

def generate_random_text(files, n, length=100):
  ngrams = get_ngrams(files, n)
  start_index = random.randint(0, len(ngrams) - 1)

  generated_text = list(ngrams[start_index])

  for _ in range(length - n):
    next_ngram_options = [ngram for ngram in ngrams if ngram[:-1] == tuple(generated_text[-(n-1):])]
    next_ngram = random.choice(next_ngram_options)
    generated_text.extend(next_ngram[-1:])

  return ' '.join(generated_text)

if __name__ == "__main__":
  main()
```

#### Obtener frecuencia de palabras más comunes

```python
# TODO
import math
import nltk
from collections import Counter

def main():
    # Load data
    print("Loading data...")
    textos_gutenberg = ["Barrica", "El_Cuervo", "Las_Campanas", "Muerte_Roja"]
    corpus = [load_text(f'{texto}.txt') for texto in textos_gutenberg]

    # Calculate TF-IDFs
    print("Calculating TF-IDFs...")
    tfidfs = calculate_tfidf(corpus)

    # Sort and get top 5 TF-IDFs for each file
    print("Computing top terms...")
    for i, tfidf_scores in enumerate(tfidfs):
        print(f"\nTop terms for {textos_gutenberg[i]}:")
        for term, score in tfidf_scores[:5]:
            print(f"    {term}: {score:.4f}")

def load_text(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def calculate_tfidf(files):
    words = set()

    for file in files:
        words.update(file.split())

    idfs = dict()

    for word in words:
        f = sum(word in file for file in files)
        idf = math.log(len(files) / f)
        idfs[word] = idf

    tfidfs = []

    for file in files:
        tfidf_scores = []
        for word in set(file.split()):
            tf = file.split().count(word)
            tfidf = tf * idfs[word]
            tfidf_scores.append((word, tfidf))
        tfidf_scores.sort(key=lambda x: x[1], reverse=True)
        tfidfs.append(tfidf_scores)

    return tfidfs

if __name__ == "__main__":
    main()
```

#### Obtener n-gramas de cada texto

```python
import nltk
from collections import Counter

def main():
    # Load data
    print("Loading data...")
    textos_gutenberg = ["Barrica", "El_Cuervo", "Las_Campanas", "Muerte_Roja"]
    corpus = [load_text(f'{texto}.txt') for texto in textos_gutenberg]

    # Compute and print n-grams for each file
    n = 3  ### Escoger el valor de n-gramas según la preferencia ###
    for i, texto in enumerate(textos_gutenberg):
        print(f"\nComputing {n}-grams for {texto}...")
        ngrams = Counter(get_ngrams([corpus[i]], n))

        print(f"\nTop {n}-grams for {texto}:")
        for ngram, freq in ngrams.most_common(5):
            print(f"{freq}: {' '.join(ngram)}")

def load_text(file_name):
    with open(file_name, 'r', encoding='utf-8') as file:
        text = file.read()
    return text

def get_ngrams(files, n):
    ngrams = []
    for file in files:
        words = file.split()
        ngrams.extend(tuple(words[i:i + n]) for i in range(len(words) - n + 1))
    return ngrams

if __name__ == "__main__":
    main()
```
